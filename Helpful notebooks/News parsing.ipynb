```python
import datetime
import gnews
import pandas as pd
import sys
from gnews import GNews
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
tweets = pd.read_csv('/content/drive/My Drive/thesis/datasets/twitter comments/Tweet.csv')
company_tweet = pd.read_csv('/content/drive/My Drive/thesis/datasets/twitter comments/Company_Tweet.csv')

tweets = pd.merge(tweets,company_tweet,on='tweet_id',how='inner')
```

```python
tweets['post_date'] = pd.to_datetime(tweets['post_date'], unit='s')
tweets['date'] = pd.to_datetime(tweets['post_date'].apply(lambda date: date.date()))
```

```python
tweets = tweets[['date']]
tweets = tweets.drop_duplicates()
```

```python
tweets = tweets.reset_index(drop = True)
```

```python
tweets['News'] = [[] for i in range(len(tweets))]
```

```python
from datetime import datetime as dtm

def format_date(date_string):
    date_obj = dtm.strptime(date_string, '%a, %d %b %Y %H:%M:%S %Z')
    return dtm.strftime(date_obj, '%Y-%m-%d')
```

```python
import datetime

for i in range(0, len(tweets)):

  year_curr = pd.to_datetime(tweets.at[i, 'date']).year
  month_curr = pd.to_datetime(tweets.at[i, 'date']).month
  day_curr = pd.to_datetime(tweets.at[i, 'date']).day

  year_next_day = pd.to_datetime(tweets.at[i, 'date'] + pd.DateOffset(1)).year
  month_next_day = pd.to_datetime(tweets.at[i, 'date'] + pd.DateOffset(1)).month
  day_next_day = pd.to_datetime(tweets.at[i, 'date'] + pd.DateOffset(1)).day


  start = datetime.date(year_curr, month_curr, day_curr)
  end = datetime.date(year_next_day, month_next_day, day_next_day)

  google_news = GNews(language='en', country='US', start_date=start, end_date=end, max_results = 50)
  rsp = google_news.get_news('Tesla')

  for article in rsp:

    date_string = article['published date']
    formatted_date = format_date(date_string)
    
    index_arr = tweets.index[tweets['date'] == formatted_date].tolist()

    if len(index_arr) > 0:

      index = index_arr[0]
      tweets.at[index, 'News'].append(article['title'])

  print(start)
```
